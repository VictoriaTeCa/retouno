```python
#importamos las librerias con las que vamos a estar trabajando durante este reto final 1
import pandas as pd
import numpy as np
import cufflinks as cf
import matplotlib.pyplot as plt
```

```python
#importamos el archivo de datos 
data = pd.read_csv("CountryTable.csv")
data
```

```python
#verificamos los valores nulos que se encuentran dentro del dataframe 
Datanulos= data.isnull()
Datanulos
```

```python
#“independence_year”: Rellenar con el valor “1900”
#“life_expectancy”: Rellenar con el valor “Media” de la columna
#“gnp_old ”: Rellenar con el valor “Mediana” de la columna
#“local_name”: Rellenar con el string “Inexistente”
#“head_of_state”: Rellenar con el string “Desconocido”
#“capital ”: Rellenar con el valor “0”
#“code2”: Rellenar con el string “NA”
data["independence_year"]=data["independence_year"].fillna(1900)
data["life_expectancy"]=data["life_expectancy"].fillna(round(data["life_expectancy"].mean(),1))
data["gnp_old"]=data["gnp_old"].fillna(round(data["gnp_old"].median(),1))
data["local_name"]=data["local_name"].fillna("Inexistente")
data["head_of_state"]=data["head_of_state"].fillna("Desconocido")
data["capital"]=data["capital"].fillna(0)
data["code2"]=data["code2"].fillna("NA")
```

```python
#comprobamos que nuestros datos esten realizados como fue solicitado en la parte superior. 
data
```

```python
#a continucación usaremos un método para identificar los outliers que se encuentran en nuestra base de datos. este proceso lo heramos con dos variables, la primera es:
#columna con variable “gnp old”. 
y=data["gnp_old"]
Limite_Superior= y.mean() + 2*y.std()
Limite_Inferior= y.mean() - 2*y.std()
outliers= data[(y>Limite_Superior)|(y<Limite_Inferior)]
outliers
```

```python
#outliers de la columna con la variable de surface_area
y=data["surface_area"]
Limite_Superior1= y.mean() + 2*y.std()
Limite_Inferior1= y.mean() - 2*y.std()
outliers_1= data[(y>Limite_Superior1)|(y<Limite_Inferior1)]
outliers_1
```

```python
# outliers de la columna con la variable de life_expectancy
y=data["life_expectancy"]
Limite_Superior2= y.mean() + 2*y.std()
Limite_Inferior2= y.mean() - 2*y.std()
outliers_2= data[(y>Limite_Superior2)|(y<Limite_Inferior2)]
outliers_2
```

```python
#outliers con la columna de la variable de population. 
y=data["population"]
Limite_Superior3= y.mean() + 2*y.std()
Limite_Inferior3= y.mean() - 2*y.std()
outliers_3= data[(y>Limite_Superior3)|(y<Limite_Inferior3)]
outliers_3
```

```python
#Aplicar método de distribución normal para identificar Outliers
y=data["gnp_old"]

percentile25=y.quantile(0.25) #Q1
percentile75=y.quantile(0.75) #Q3
iqr= percentile75 - percentile25

Limite_Superior_iqr= percentile75 + 1.3*iqr
Limite_Inferior_iqr= percentile25 - 1.3*iqr
outliers_iqr= data[(y>Limite_Superior_iqr)|(y<Limite_Inferior_iqr)]
outliers_iqr
```
